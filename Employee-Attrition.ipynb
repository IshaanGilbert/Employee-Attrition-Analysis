```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
import random

# Step 1: Load the dataset (mock dataset for Employee Attrition Analysis)
data = {
    "EmployeeID": [i for i in range(1, 101)],
    "Age": [round(random.uniform(22, 60)) for _ in range(100)],
    "Gender": [random.choice(["Male", "Female"]) for _ in range(100)],
    "Department": [random.choice(["Sales", "HR", "IT", "Finance"]) for _ in range(100)],
    "YearsAtCompany": [round(random.uniform(1, 20)) for _ in range(100)],
    "JobSatisfaction": [random.choice(["Low", "Medium", "High", "Very High"]) for _ in range(100)],
    "MonthlyIncome": [round(random.uniform(3000, 20000), 2) for _ in range(100)],
    "Attrition": [random.choice(["Yes", "No"]) for _ in range(100)],
}

# Create the DataFrame
attrition_df = pd.DataFrame(data)

# Display the first few rows
print(attrition_df.head())
```

```python
# Create the DataFrame
attrition_df = pd.DataFrame(data)
```

```python
# Step 2: Data Cleaning
print("\nStep 2: Data Cleaning")
# Checking for missing values
print("\nMissing Values:")
print(attrition_df.isnull().sum())
```

```python
# Step 3: Exploratory Data Analysis (EDA)
print("\nStep 3: Exploratory Data Analysis")
```

```python
# 1. Attrition Rate
attrition_rate = attrition_df['Attrition'].value_counts(normalize=True) * 100
print("\nAttrition Rate (%):")
print(attrition_rate)
```

```python
# 2. Attrition by Department
department_attrition = attrition_df[attrition_df['Attrition'] == "Yes"]['Department'].value_counts()
print("\nAttrition by Department:")
print(department_attrition)
```

```python
# 3. Attrition by Job Satisfaction
satisfaction_attrition = attrition_df.groupby('JobSatisfaction')['Attrition'].value_counts(normalize=True).unstack()["Yes"] * 100
print("\nAttrition by Job Satisfaction (%):")
print(satisfaction_attrition)
```

```python
# Step 4: Data Visualization
print("\nStep 4: Data Visualization")
sns.set_style("whitegrid")
```

```python
# Pie chart: Attrition rate
plt.figure(figsize=(6, 6))
attrition_rate.plot(kind='pie', autopct='%1.1f%%', colors=['gold', 'lightblue'], startangle=90)
plt.title('Attrition Rate')
plt.ylabel('')
plt.show()
```

```python
# Bar chart: Attrition by department
plt.figure(figsize=(8, 5))
department_attrition.plot(kind='bar', color='skyblue')
plt.title('Attrition by Department')
plt.ylabel('Number of Employees')
plt.xlabel('Department')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar chart: Attrition by job satisfaction
plt.figure(figsize=(8, 5))
satisfaction_attrition.sort_values().plot(kind='bar', color='orange')
plt.title('Attrition by Job Satisfaction')
plt.ylabel('Attrition Rate (%)')
plt.xlabel('Job Satisfaction')
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 5: Insights and Recommendations
print("\nStep 5: Insights and Recommendations")
print("\nInsights:")
print("1. Overall attrition rate is", round(attrition_rate["Yes"], 2), "%.")
print("2. Departments with the highest attrition are:")
print(department_attrition.head())
print("3. Employees with 'Low' job satisfaction have significantly higher attrition rates.")
```
